## Classrooms

```python
classroom = pd.read_csv('classrooms.csv')
classroom = classroom.fillna(0)
locations = pd.read_csv('unique_building.csv')
classroom = classroom.merge(locations)
classroom.head(5)
```

```python
classroom_sum = classroom.groupby(['Building', 'Latitude', 'Longitude']).sum().sort_values('Building')
classroom_count = classroom.groupby(['Building', 'Latitude', 'Longitude']).count().sort_values('Building')
```

```python
classroom_sum['Total Rooms'] = classroom_count.iloc[:,0]
cc = classroom_sum[['Course Capture', 'Total Rooms']].sort_values('Course Capture', ascending = False).reset_index()
cc['%'] = cc['Course Capture'] / cc['Total Rooms']
cc.to_csv('mapping.csv')
cc.head(5)
```

```python
seats = classroom_sum[['Course Capture', 'Seats']].sort_values('Course Capture', ascending = False).reset_index()
seats.to_csv('mapping_seats.csv')
seats.head(5)
```

```python
classroom.groupby(['Building']).count()['Classroom Number'].sort_values(ascending = False).reset_index().head(5)
```

```python
classroom.groupby(['Building']).sum()['Seats'].sort_values(ascending = False).reset_index().head(5)
```

## Department

```python
fall2016 = pd.read_csv('F2016.csv')
fall2016.head(5)
```

```python
#number of classes by building
fall2016.groupby(['Building']).count()['Subject'].reset_index().rename(
    columns = {'Subject':'Count'}).sort_values('Count', ascending = False)
```

```python
## seats usage by department
fall2016['c_n'] = fall2016['Building'] + fall2016['Classroom']
classroom['c_n'] = classroom['Building'] + classroom['Classroom Number']
by_class = fall2016.merge(classroom, on = 'c_n')

dept_seats = by_class.groupby('Subject')['Seats'].sum().reset_index().sort_values('Seats', ascending = False)
dept_seats.head(5)
```

```python
### Joining classes and classroom dataset

combined = classroom.merge(fall2016, on ='c_n').drop(
    ['Building_y', 'Classroom', 'Latitude', 'Longitude', 'Unnamed: 0', 'c_n'], axis = 1).rename(columns = {'Building_x':'Building'})
combined.head(5)
```

```python
#CCC rate by subject
top_majors = ['ECON', 'STAT', 'HISTORY', 'COMPSCI', 'UGBA', 'POLSCI', 'ESPM', 'PSYCH','CHMENG', 'COGSCI', 'CHMENG',
              'MATH', 'MCELLBI', 'ENGLISH', 'INTEGBI', 'CHEM', 'ELENG', 'Physics', 'SOCIOL', 'PBHLTH', 'LINGUIS', 
              'MEDIAST']

dept_score = combined.groupby('Subject').mean()['Course Capture'].reset_index().rename(
    columns = {'Course Capture':'%'}).sort_values('%', ascending = False)
dept_score[dept_score['Subject'].isin(top_majors)]
```

```python
### AverageCCC Percentage
np.mean(dept_score['%'])
```

```python
# CCC numbers by subject
dept_ccroom_sum = combined.groupby('Subject').sum()['Course Capture'].reset_index().rename(
    columns = {'Course Capture':'Count'}).sort_values('Count', ascending = False)

dept_ccroom_sum[dept_ccroom_sum['Subject'].isin(top_majors)]
```

```python
### look into one department
combined[combined['Subject'] == 'POLSCI'].groupby(['Building', 'Course Capture']).count()['Classroom Number'].reset_index().sort_values(
    'Building').rename(columns = {'Classroom Number': 'Classrooms Used'})

combined[combined['Subject'] == 'POLSCI'].groupby(['Building', 'Classroom Number']).count()['Seats'].reset_index()
```

```python
## CCC seats by department
dept_seats_cc = pd.pivot_table(by_class, values='Seats', index = 'Subject', columns = 'Course Capture', 
                               aggfunc = np.sum).reset_index().rename(columns = {0: 'No', 1: 'Yes'}).fillna(0)
dept_seats_cc['cc_ratio'] = dept_seats_cc['Yes'] / (dept_seats_cc['Yes'] + dept_seats_cc['No'])
dept_seats_cc[dept_seats_cc['Subject'].isin(top_majors)].sort_values(['cc_ratio','Yes'], ascending = False)
```

```python
np.mean(dept_seats_cc['cc_ratio'])
```

## Mapping buildings by CCC seats

```python
build_seats_cc = pd.pivot_table(classroom, values='Seats', index = 'Building', columns = 'Course Capture', 
                               aggfunc = np.sum).reset_index().rename(columns = {0: 'No', 1: 'Yes'}).fillna(0)
build_seats_cc['cc_ratio'] = build_seats_cc['Yes'] / (build_seats_cc['Yes'] + build_seats_cc['No'])
#build_seats_cc[build_seats_cc['Building_x'].isin(top_majors)].sort_values(['cc_ratio','Yes'], ascending = False)
build_seats_cc_final = build_seats_cc.merge(seats[['Building','Latitude','Longitude']], left_on = 'Building', right_on = 'Building')[[
    'Building','Latitude','Longitude', 'No', 'Yes', 'cc_ratio']]
build_seats_cc_final.to_csv('b_seats_cc.csv')
```

```python
STEM = ['ENGIN', 'ESPM',
       'MATH', 'MEDIAST', 
       'CHEM', 'ENVECON', 
       'MCELLBI',  'ECON',  'PLANTBI', 
        'BIOENG', 'ELENG', 'INDENG',
        'STAT', 'MECENG', 'CHMENG', 'EPS', 'GEOG',
       'INTEGBI', 'CIVENG', 'COGSCI', 
       'Physics ', 'COMPSCI', 'ENERES', 'NUSCTX',
       'PBHLTH', 
        'PSYCH', 
        'BIOLOGY', 
       'NUCENG', 'EECS', 'INFO', 'DESINV', 'LDARCH',
         'ASTRON', 'MATSCI', 'CMPBIO',
       'AEROSPC', 'ENVDES']

combined['STEM'] = ['Yes' if i in STEM else 'No' for i in combined['Subject']]
```

```python
STEM_score = combined.groupby('STEM').mean()['Course Capture'].reset_index().rename(
    columns = {'Course Capture':'%'}).sort_values('%', ascending = False)
STEM_score
```

```python
STEM_seats_cc = pd.pivot_table(combined, values='Seats', index = 'STEM', columns = 'Course Capture', 
                               aggfunc = np.sum).reset_index().rename(columns = {0: 'No', 1: 'Yes'}).fillna(0)
STEM_seats_cc['cc_ratio'] = STEM_seats_cc['Yes'] / (STEM_seats_cc['Yes'] + STEM_seats_cc['No'])
STEM_seats_cc
```

## By Colleges

```python
### By majors

coe = ['AEROSPC', 'BIOENG', 'CIVENG', 'EECS', 'ENGIN', 'INDENG', 'MATSCI',
      'NUCENG', 'ELENG', 'MECENG']
coc =['CHEM', 'CHMENG']
ced =['LDARCH', 'ENVDES']
cnr = ['ENERES', 'ENVECON' ,'ESPM', 'NUSCTX', 'PLANTBI']
haas = ['UGBA']

def college(s):
    if s in coe:
        return 'CoE'
    elif s in coc:
        return 'CoC'
    elif s in ced:
        return 'CED'
    elif s in cnr:
        return 'CNR'
    elif s in haas:
        return 'Haas'
    else:
        return 'L&S'
    
combined['college'] = [college(i) for i in combined['Subject']]
```

```python
college_department_score = combined.groupby('college').mean()['Course Capture'].reset_index().rename(
    columns = {'Course Capture':'%'}).sort_values('%', ascending = False)
college_department_score.sort_values('college')
```

```python
college_department_seats_cc = pd.pivot_table(combined, values='Seats', index = 'college', columns = 'Course Capture', 
                               aggfunc = np.sum).reset_index().rename(columns = {0: 'No', 1: 'Yes'}).fillna(0)
college_department_seats_cc['cc_ratio'] = college_department_seats_cc['Yes'] / (college_department_seats_cc['Yes'] 
                                                                                + college_department_seats_cc['No'])
college_department_seats_cc
```
